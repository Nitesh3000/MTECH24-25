#Image Enhancement

##Original Image

The image that i chose to enhance is of dimentions (480,640)

It has Mean Intensity of 29.77 and standard deviation of 52.76


The histogram (displayed in the output) is the skewed histogram and is biased toward the lower end of the intensity scale indicating dark image.

Also the histogram shows lack of wide range of pixel values indicating that it is a low contrast image

This infers that image is relatively dark (as seen below in the output), with a low mean intensity and moderate standard deviation(low contrast)

No information can be grabbed from the image that is below the mesh/window as it is completly dark. Same for the section the is above and to the right of the mesh.

##Histogram Equalized Image

As seen in the graph below, the histogram of the equalized image shows a uniform distribution of pixel intensity values across the entire intensity range as compared to that of the original image

The peak of the histogram is more centered, and the distribution is more balanced (improved image contrast than that of the original image)

Mean intensity increased to 131.87 as an outcome of equalization process(brightening the image to enhance visibility)

Standard deviation of the image also increased 70.61 so that the there is an effective increase in the spread og the pixel values, improving image contrast.

So, in this image, the details and features in the image are more visible, providing a clearer representation of its contents.

The parts below the window/mesh are clearly visible.

But the border that occurs in the section below the mesh/window is not smooth due to high contrast(patches of black can be seen)

So to remove this contrast adjustment is done

##Contrast Adjusted Image

By adjusting the contrast of the image changing different alpha and beta values in the code provided below, the best result that i got was when i set alpha=0.9 and beta=30.

alpha controlled the overall brightness nad beta controla the darkness.

Resulted adjusted image is a good balance between the brighness and the contrast.

the adjusted image is relatively bright, and the contrast has improved preserving the details and the features. The edges below is a little smooth like due to contrast adjustment and removal of black patches is visible in the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread('/content/drive/MyDrive/Image for Image Processing/im1.jpg', cv2.IMREAD_GRAYSCALE)

#doing histogram equalization of the original image
equalized_image = cv2.equalizeHist(image)

alpha = 0.9 #for adjusting Brightness

beta = 30#for adjusting darkness

#the below line performs a pixel-wise intensity scaling and offset adjustment in an image.
contrast_adjusted_image = cv2.convertScaleAbs(equalized_image, alpha=alpha, beta=beta)

#below lines are for the calculation of histogram of the images
hist_original=cv2.calcHist([image],[0],None,[256],[0, 256])
hist_equalized=cv2.calcHist([equalized_image],[0],None,[256],[0, 256])
hist_adjusted=cv2.calcHist([contrast_adjusted_image],[0],None,[256],[0, 256])

#below lines are regarding the dimensions, mean intensity and standard deviation of the original image
print("Information about the Original Image:")
print(f"Dimensions: {image.shape}")
print(f"Mean Intensity: {np.mean(image)}")
print(f"Standard Deviation: {np.std(image)}")

#below lines are regarding the mean intensity and standard deviation of the equalized image
print("\nInformation about the Equalized Image:")
print(f"Mean Intensity: {np.mean(equalized_image)}")
print(f"Standard Deviation: {np.std(equalized_image)}")

#below lines are regarding the mean intensity and standard deviation of adjusted image
print("\nInformation about the Adjusted Image:")
print(f"Mean Intensity: {np.mean(contrast_adjusted_image)}")
print(f"Standard Deviation: {np.std(contrast_adjusted_image)}")# Plot the histograms

#Plotting the histogram of the three images(original, equalized and histogram)
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title('Original Image Histogram')
plt.plot(hist_original,color='b')
plt.xlim([0, 256])

plt.subplot(1, 3, 2)
plt.title('Equalized Image Histogram')
plt.plot(hist_equalized, color='b')
plt.xlim([0, 256])

plt.subplot(1, 3, 3)
plt.title('Adjusted Image Histogram')
plt.plot(hist_adjusted, color='b')
plt.xlim([0, 256])

plt.show()

#Plotting the original, equalized and adjusted images
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Equalized Image')
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Adjusted Image')
plt.imshow(cv2.cvtColor(contrast_adjusted_image, cv2.COLOR_BGR2RGB), cmap='gray')
plt.axis('off')

plt.show()


```
