```python
# Image smoothing using low-pass filter

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image file
image = cv2.imread('/content/drive/MyDrive/Image for Image Processing/desktop-wallpaper-awesome-top-awesome-backgrounds-25-terrific.jpg',cv2.IMREAD_GRAYSCALE)

# Define a 3x3 averaging kernel for image smoothing
kernel = np.array([[1, 1, 1],[1, 1, 1],[1, 1, 1]]) / 9

# Get the dimensions of the image
height, width = image.shape

# Create a blank image to store the smoothed result
smoothed_image = np.zeros_like(image)

# Apply the kernel to the grayscale image
for y in range(1, height - 1):
    for x in range(1, width - 1):
        smoothed_image[y, x] = np.sum(
            image[y - 1:y + 2, x - 1:x + 2] * kernel)

plt.subplot(121)
plt.imshow(image, cmap = 'gray')
plt.title('Input Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(smoothed_image, cmap = 'gray')
plt.title('Smoothed Image')
plt.axis('off')
```

```python
# High pass filter (Approach 1)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image file
image = cv2.imread('/content/drive/MyDrive/Image for Image Processing/desktop-wallpaper-awesome-top-awesome-backgrounds-25-terrific.jpg',cv2.IMREAD_GRAYSCALE)

# Define a 3x3 Gaussian kernel for low-pass filtering
kernel_size = (3, 3)
sigma = 1.0
low_pass_image = cv2.GaussianBlur(image, kernel_size, sigma)

# Calculate the high-pass image by subtracting the low-pass from the original
high_pass_image = image - low_pass_image

# Optionally, normalize the high-pass image to enhance details
#high_pass_image = cv2.normalize(high_pass_image, None, 0, 255, cv2.NORM_MINMAX)

plt.subplot(121)
plt.imshow(image, cmap = 'gray')
plt.title('Input Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(high_pass_image, cmap = 'gray')
plt.title('High Frequency Regions')
plt.axis('off')
```

```python
# Using laplacian
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image file
image = cv2.imread('/content/drive/MyDrive/Image for Image Processing/img_bgr.jpg',cv2.IMREAD_GRAYSCALE)

# Define a Laplacian kernel for high-pass filtering
laplacian_kernel = np.array([[0, 1, 0],
                              [1, -4, 1],
                              [0, 1, 0]])

# Apply the Laplacian kernel to the grayscale image
high_pass_image = cv2.filter2D(image, -1, laplacian_kernel)

# Optionally, normalize the high-pass image to enhance details
high_pass_image = cv2.normalize(high_pass_image, None, 0, 255, cv2.NORM_MINMAX)

plt.subplot(121)
plt.imshow(image, cmap = 'gray')
plt.title('Input Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(high_pass_image, cmap = 'gray')
plt.title('Laplacian Enhanced Image')
plt.axis('off')
```
