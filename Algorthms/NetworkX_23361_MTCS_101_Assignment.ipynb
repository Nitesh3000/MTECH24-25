NetworkX provides different types of graphs, including directed graphs (DiGraph), undirected graphs (Graph), and others

Undirected Graph

```python
import networkx as nx
G = nx.Graph() # Create an empty undirected graph
```

Add nodes

```python
G.add_node(1)
G.add_nodes_from([2, 3])
```

Add edges

```python
G.add_edge(1, 2)
G.add_edges_from([(2, 3), (1, 3)])
```

Visualization

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```

NetworkX offers a wide range of graph algorithms, such as shortest path..

```python
shortest_path = nx.shortest_path(G, source=1, target=3)
print("Shortest Path:", shortest_path)
```

You can analyze various properties of the graph, like degree distribution

```python
print("Node Degrees:", dict(G.degree()))
```

Directed Graph

```python
G = nx.DiGraph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
```

```python
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 1)
```

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold', arrows=True)
plt.show()
```

```python
shortest_path = nx.shortest_path(G, source=1, target=3)
print("Shortest Path:", shortest_path)
```

```python
print("Node Degrees:", dict(G.degree()))
```

```python
for node in G.nodes():
    in_degree = G.in_degree(node)
    out_degree = G.out_degree(node)
    print(f"Node {node}: In-Degree = {in_degree}, Out-Degree = {out_degree}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

G.add_edge(1, 2, weight=1)
G.add_edge(1, 3, weight=4)
G.add_edge(2, 3, weight=2)
G.add_edge(2, 4, weight=7)
G.add_edge(3, 4, weight=3)
G.add_edge(3, 5, weight=5)
G.add_edge(4, 5, weight=2)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold', node_size=600)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Initial Directed Graph")
plt.show()

# Dijkstra's algorithm
shortest_path = nx.shortest_path(G, source=1, target=5, weight='weight')
shortest_path_length = nx.shortest_path_length(G, source=1, target=5, weight='weight')

print(f"Shortest Path from Node 1 to Node 5: {shortest_path}")
print(f"Shortest Path Length: {shortest_path_length}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

G.add_edge(1, 2, weight=1)
G.add_edge(1, 3, weight=4)
G.add_edge(2, 3, weight=2)
G.add_edge(2, 4, weight=7)
G.add_edge(3, 4, weight=3)
G.add_edge(3, 5, weight=5)
G.add_edge(4, 5, weight=2)

UG = G.to_undirected()

MST = nx.minimum_spanning_tree(UG, weight='weight')

print("Minimum Cost Spanning Tree Edges:")
for edge in MST.edges(data=True):
    print(edge)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(UG, seed=42)
labels = nx.get_edge_attributes(UG, 'weight')
nx.draw(UG, pos, with_labels=True, node_color='lightblue', font_weight='bold', node_size=600)
nx.draw_networkx_edge_labels(UG, pos, edge_labels=labels)
nx.draw_networkx_edges(MST, pos, edgelist=MST.edges(), edge_color='green', width=2)
plt.title("Minimum Cost Spanning Tree")
plt.show()
```
