##Heapify

```python
import random
```

```python
size=10000
A=[]
for _ in range(size):
  A.append(random.randint(0,1000))
print(A)
n=len(A)
B=A.copy()
```

```python
#bottomUp
def Alg_BottomUp(A):

  count=0
  for i in range((n//2)-1,-1,-1):
    for j in range(i,(n//2)):
      small=j
      if (2*j+1)<n:
        if A[2*j+1]<A[small]:
          small=2*j+1

      if (2*j+2)<n:
        if A[2*j+2]<A[small]:
          small=2*j+2

      if j!=small:
        A[j],A[small]=A[small],A[j]
        j=small
        count=count+1
  return count,n,A
```

```python
def Alg_topDown(B):
    n = len(B)
    count = 0

    for i in range(n):
        j = i

        while j > 0:
            parent = (j - 1) // 2

            if B[j] < B[parent]:
                B[j], B[parent] = B[parent], B[j]
                count += 1
                j = parent
            else:
                break

    return count,n,B
```

```python
print(Alg_topDown(B))
print(Alg_BottomUp(A))
```

##Scalar Multiplication for Grayscale Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image_path = '/content/drive/MyDrive/Image for Image Processing/desktop-wallpaper-awesome-top-awesome-backgrounds-25-terrific.jpg'
alpha = 3.0

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
print("image: ",image)
rows, cols = image.shape
# Create an empty result array with the same dimensions as the input image
result = np.zeros((rows, cols), dtype=np.float32)
# Initialize an array to store the computed transformation values
zeroo=[0]

for i in range(1,256):
  zeroo.append((alpha+zeroo[i-1]))

for i in range(rows):
    for j in range(cols):
        result[i, j]=zeroo[int(image[i,j])]
print("result is: ",result)

# Ensure the values are within the [0, 255] range
result = np.clip(result, 0, 255)
print(result)

result = result.astype(np.uint8)

plt.imshow(image,cmap='gray')
plt.title('original Image')
plt.axis('off')
plt.show()

plt.imshow(result,cmap='gray')
plt.title('Scalar Multiplication')
plt.axis('off')
plt.show()
```
