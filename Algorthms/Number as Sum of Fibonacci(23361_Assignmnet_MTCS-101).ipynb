```python
a = int(input('Give input: '))

def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
fib=(list(fib(a)))
print(fib)
```

```python
a=int(input('Input the number which you want to represent as sum of fibonacci series: '))
```

```python
x=a
sum=0
listt=[]
def fibnumber(fib,a,x,sum,listt):
  if a == 0:
    return listt
  if fib[len(fib)-1]==a:
    return fib[len(fib)-1]
  for i in range(len(fib)-1):
   if fib[i+1]>a:
    sum=sum+fib[i]
    listt.append(fib[i])
    return fibnumber(fib,x-sum,x,sum,listt)
```
