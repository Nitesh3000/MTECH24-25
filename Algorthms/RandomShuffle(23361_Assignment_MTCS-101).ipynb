```python
import random
Comp=[]
B=[]
for _ in range(6):
  B.append(random.randint(0,1000))
for j in range(10000):
  A=B.copy()
  n=len(B)

  for i in range(n-1):
          k=random.randint(i,n-1)
          A[k],A[i]=A[i],A[k]
  Comp.append(A)
my_set = {tuple(x) for x in Comp}
print(my_set)
print(len(my_set))
```

```python
import random
Comp=[]
my_set=set()
B=[]
for _ in range(10):
  B.append(random.randint(0,10000))
for j in range(1000000):
  A=B.copy()
  n=len(A)
  mapping = {element: random.randint(0, n**3) for element in A}
  sorted_mapping = dict(sorted(mapping.items(), key=lambda item: item[1]))
  shuffled_array = list(sorted_mapping.keys())
  #comp.append(A)
  my_set.add(tuple(shuffled_array))
print(len(my_set))
```
